```python
import pandas as pd 
import spacy 
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
content = """
The Indian Space Research Organisation (ISRO) successfully launched the Gaganyaan test mission from the Satish Dhawan Space Centre in Sriharikota on Saturday. 
Prime Minister Narendra Modi congratulated the scientists for their efforts and said the mission marked a major milestone in India’s human spaceflight program. 
ISRO Chairman S. Somanath stated that the crew module performed as expected and all safety parameters were met during the test. 

Meanwhile, Union Home Minister Amit Shah chaired a high-level meeting in New Delhi to review internal security arrangements ahead of the upcoming general elections. 
Senior officials from the Intelligence Bureau, National Investigation Agency, and Delhi Police were present at the meeting. 

On the international front, US President Joe Biden held talks with UK Prime Minister Rishi Sunak at the White House. 
The leaders discussed the Russia-Ukraine war, global economic challenges, and climate change commitments made under the Paris Agreement. 

In the business sector, Tesla CEO Elon Musk announced plans to set up a manufacturing plant in Gujarat, creating over 5,000 jobs. 
Musk said the facility would support India’s push towards electric vehicles and sustainable energy solutions.
"""

doc = nlp(content)

for ent in doc.ents:
    print(ent.text, ent.label_, ent.lemma_)
```

```python
from spacy import displacy
displacy.render(doc, style="ent")
```
